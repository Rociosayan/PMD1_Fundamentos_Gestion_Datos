### 📍 Paso 1: Análisis de características de los datos del proyecto (las 4 V's)

````python
# ============================================================
# PASO 1: Análisis de las 4 V's del proyecto
# ============================================================
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

# Conectar a la BD del proyecto
# Opción A: Drive → DB_PATH = '/content/drive/MyDrive/proyecto/proyecto.db'
# Opción B: BD de ejemplo en memoria
DB_PATH = 'proyecto_arq.db'
conn = sqlite3.connect(DB_PATH)
cursor = conn.cursor()

# Crear tablas de ejemplo (reemplazar con tu BD real)
cursor.executescript('''
    CREATE TABLE IF NOT EXISTS clientes (
        id_cliente INTEGER PRIMARY KEY, nombre TEXT, email TEXT,
        ciudad TEXT, fecha_reg TEXT
    );
    CREATE TABLE IF NOT EXISTS productos (
        id_producto INTEGER PRIMARY KEY, nombre TEXT,
        categoria TEXT, precio REAL, stock INTEGER
    );
    CREATE TABLE IF NOT EXISTS ventas (
        id_venta INTEGER PRIMARY KEY, id_cliente INTEGER,
        id_producto INTEGER, cantidad INTEGER,
        fecha_venta TEXT, total REAL
    );
''')
conn.commit()

# Analizar las 4 V's
tablas = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table'", conn)
print('📊 ANÁLISIS DE LAS 4 V\' DEL PROYECTO')
print('='*50)

volumen_total = 0
for t in tablas['tabla'] if 'tabla' in tablas.columns else tablas['name']:
    count = pd.read_sql(f'SELECT COUNT(*) as n FROM {t}', conn).iloc[0,0]
    cols  = len(pd.read_sql(f'SELECT * FROM {t} LIMIT 1', conn).columns)
    volumen_total += count
    print(f'  Tabla {t:15s}: {count:6,} registros × {cols} columnas')

print(f'\n  VOLUMEN TOTAL: {volumen_total:,} registros')
print(f'  VARIEDAD: Datos estructurados (SQLite)')
print(f'  VELOCIDAD: Batch (actualización periódica)')
print(f'  VERACIDAD: Pendiente de análisis de calidad (Semana 13)')
````

### 📍 Paso 5: Reporte automático de estadísticas del proyecto

Este reporte simula la **capa de servicio** de tu arquitectura: consultas analíticas que generan valor de negocio.

````python
# ============================================================
# PASO 5: Reporte automático con estadísticas (capa de servicio)
# ============================================================
import warnings
warnings.filterwarnings('ignore')

# Insertar datos de ejemplo si la BD está vacía
cursor.executemany('INSERT OR IGNORE INTO clientes VALUES (?,?,?,?,?)',
    [(i, f'Cliente {i}', f'c{i}@mail.com', ['Lima','Arequipa','Cusco','Trujillo'][i%4],
      f'2026-0{(i%3)+1}-{(i%28)+1:02d}') for i in range(1,51)])
cursor.executemany('INSERT OR IGNORE INTO productos VALUES (?,?,?,?,?)',
    [(i, f'Producto {i}', ['Electrónica','Ropa','Alimentos','Hogar'][i%4],
      round(50 + i*15.5, 2), 100-i) for i in range(1,21)])
cursor.executemany('INSERT OR IGNORE INTO ventas VALUES (?,?,?,?,?,?)',
    [(i, (i%50)+1, (i%20)+1, (i%5)+1, f'2026-0{(i%3)+1}-{(i%28)+1:02d}',
      round(((i%20)+1)*15.5 * ((i%5)+1), 2)) for i in range(1,101)])
conn.commit()

# Reporte de estadísticas
print('📊 REPORTE AUTOMÁTICO — CAPA DE SERVICIO')
print('='*55)
print(f'Generado: {datetime.now().strftime("%Y-%m-%d %H:%M")}\n')

# KPIs principales
kpis = [
    ('Total clientes'    , 'SELECT COUNT(*) FROM clientes'),
    ('Total productos'   , 'SELECT COUNT(*) FROM productos'),
    ('Total ventas'      , 'SELECT COUNT(*) FROM ventas'),
    ('Venta total (PEN)' , 'SELECT ROUND(SUM(total),2) FROM ventas'),
    ('Ticket promedio'   , 'SELECT ROUND(AVG(total),2) FROM ventas'),
]
for label, query in kpis:
    val = pd.read_sql(query, conn).iloc[0,0]
    print(f'  {label:<22}: {val:>12,.2f}' if isinstance(val, float) else f'  {label:<22}: {val:>12,}')

# Top 5 ciudades por ventas
print('\n🏙️  TOP CIUDADES POR VENTAS:')
top_ciudades = pd.read_sql('''
    SELECT c.ciudad, COUNT(v.id_venta) as ventas, ROUND(SUM(v.total),2) as monto
    FROM ventas v JOIN clientes c ON v.id_cliente = c.id_cliente
    GROUP BY c.ciudad ORDER BY monto DESC LIMIT 5
''', conn)
print(top_ciudades.to_string(index=False))
````

````python
# Visualizaciones del reporte
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('📊 Dashboard — Proyecto Integrador (Capa de Servicio)', fontsize=14, fontweight='bold')

# Gráfico 1: Ventas por ciudad
top_ciudades.plot(kind='bar', x='ciudad', y='monto', ax=axes[0],
                  color='steelblue', legend=False)
axes[0].set_title('Ventas por Ciudad (PEN)')
axes[0].set_xlabel('Ciudad')
axes[0].set_ylabel('Monto total')
axes[0].tick_params(axis='x', rotation=45)

# Gráfico 2: Top 5 productos más vendidos
top_prod = pd.read_sql('''
    SELECT p.nombre, SUM(v.cantidad) as unidades
    FROM ventas v JOIN productos p ON v.id_producto = p.id_producto
    GROUP BY p.nombre ORDER BY unidades DESC LIMIT 5
''', conn)
axes[1].barh(top_prod['nombre'], top_prod['unidades'], color='coral')
axes[1].set_title('Top 5 Productos más Vendidos')
axes[1].set_xlabel('Unidades vendidas')

# Gráfico 3: Distribución por categoría
cat = pd.read_sql('''
    SELECT p.categoria, COUNT(v.id_venta) as ventas
    FROM ventas v JOIN productos p ON v.id_producto = p.id_producto
    GROUP BY p.categoria
''', conn)
axes[2].pie(cat['ventas'], labels=cat['categoria'], autopct='%1.1f%%',
            colors=['#4C72B0','#DD8452','#55A868','#C44E52'])
axes[2].set_title('Ventas por Categoría (%)')

plt.tight_layout()
plt.savefig('reporte_dashboard_proyecto.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Dashboard guardado como imagen')
````

### 📍 Paso 6: Análisis de escalabilidad y arquitectura futura

````python
# ============================================================
# PASO 6: Proyección de escalabilidad del proyecto
# ============================================================

# Parámetros de crecimiento (ajusta según tu proyecto)
registros_iniciales = 100          # Registros actuales en tabla ventas
crecimiento_mensual = 0.20         # 20% de crecimiento mensual
limite_sqlite_mb    = 2000         # Límite práctico de SQLite: ~2GB
bytes_por_registro  = 500          # Estimación promedio de bytes por registro

meses = list(range(0, 37))  # 3 años
registros_proyectados = [registros_iniciales * (1 + crecimiento_mensual)**m for m in meses]
tamanio_mb = [r * bytes_por_registro / (1024**2) for r in registros_proyectados]

# Encontrar mes límite
mes_limite = next((i for i, mb in enumerate(tamanio_mb) if mb > limite_sqlite_mb), None)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('📈 Proyección de Escalabilidad — Proyecto', fontsize=13, fontweight='bold')

ax1.plot(meses, registros_proyectados, color='steelblue', linewidth=2)
ax1.set_title('Crecimiento de Registros (36 meses)')
ax1.set_xlabel('Mes')
ax1.set_ylabel('Número de registros')
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))
ax1.grid(True, alpha=0.3)

ax2.plot(meses, tamanio_mb, color='coral', linewidth=2)
ax2.axhline(y=limite_sqlite_mb, color='red', linestyle='--', label=f'Límite SQLite ({limite_sqlite_mb}MB)')
if mes_limite:
    ax2.axvline(x=mes_limite, color='orange', linestyle=':', label=f'Mes {mes_limite}: migrar a nueva arquitectura')
ax2.set_title('Crecimiento del Tamaño de BD')
ax2.set_xlabel('Mes')
ax2.set_ylabel('Tamaño (MB)')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f'📊 RESUMEN DE ESCALABILIDAD:')
print(f'  Registros actuales: {registros_iniciales:,}')
print(f'  Registros en 12 meses: {registros_proyectados[12]:,.0f}')
print(f'  Registros en 36 meses: {registros_proyectados[36]:,.0f}')
if mes_limite:
    print(f'  ⚠️  SQLite alcanza su límite en el mes {mes_limite}')
    print(f'  → Recomendación: migrar a PostgreSQL/Snowflake en el mes {mes_limite}')
else:
    print('  ✅ SQLite es suficiente para 36 meses con este crecimiento')

conn.close()
````
