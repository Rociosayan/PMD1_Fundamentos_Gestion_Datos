## ¿Qué es una variable?

Una variable es cada columna de un conjunto de datos.  
Representa una característica que se mide u observa en cada registro (fila).

---

# Clasificación de variables

| Tipo | Subtipo | Descripción | Ejemplo |
|:---|:---|:---|:---|
| Cuantitativa | Continua | Toma cualquier valor numérico en un rango | Peso: 65.3 kg, temperatura: 36.7 °C |
| Cuantitativa | Discreta | Solo valores enteros contables | Número de hijos: 0, 1, 2, 3 |
| Cualitativa | Nominal | Categorías sin orden entre sí | Color, país, género |
| Cualitativa | Ordinal | Categorías con orden lógico | Nivel: básico < medio < superior |

---

> **Importante:** Un `int64` no siempre es cuantitativo.

# Estructura de la base de datos

La base de datos tiene tres tablas relacionadas:

```text
clientes     (cliente_id PK, nombre NOT NULL, ciudad NOT NULL)

medicamentos (medicamento_id PK, nombre NOT NULL UNIQUE, categoria NOT NULL, precio NOT NULL)

ventas       (venta_id PK, cliente_id FK -> clientes,
              medicamento, categoria, ciudad,
              cantidad NOT NULL, monto NOT NULL, fecha NOT NULL)
```

---

## Significado de las restricciones

| Abreviatura / Restricción | Significado |
|:---|:---|
| `PK` | Clave Primaria |
| `FK` | Clave Foránea |
| `NOT NULL` | Campo obligatorio |
| `UNIQUE` | No permite valores repetidos |

# Celda de preparación


````python

import sqlite3
import pandas as pd
import os

DB_NAME = "farmacia.db"

# Eliminar la base anterior si ya existe
if os.path.exists(DB_NAME):
    os.remove(DB_NAME)

# Crear conexión
conexion = sqlite3.connect(DB_NAME)

# Crear cursor
cursor = conexion.cursor()

print("Base de datos creada correctamente.")

# ═══════════════════════════════════════════════════════════════════════════
# CREAR TABLAS
# ═══════════════════════════════════════════════════════════════════════════

# Tabla 1: clientes
#   cliente_id → PRIMARY KEY (identificador único de cada cliente)
#   nombre, ciudad → NOT NULL (campos obligatorios)
cursor.execute("""
CREATE TABLE clientes (
    cliente_id  INTEGER  PRIMARY KEY,
    nombre      TEXT     NOT NULL,
    ciudad      TEXT     NOT NULL
);
""")

# Tabla 2: medicamentos
#   medicamento_id → PRIMARY KEY
#   nombre → NOT NULL y UNIQUE (no puede haber dos medicamentos con el mismo nombre)
#   precio → NOT NULL
cursor.execute("""
CREATE TABLE medicamentos (
    medicamento_id  INTEGER  PRIMARY KEY,
    nombre          TEXT     NOT NULL UNIQUE,
    categoria       TEXT     NOT NULL,
    precio          REAL     NOT NULL
);
""")

# Tabla 3: ventas
#   venta_id → PRIMARY KEY
#   cliente_id → FOREIGN KEY que apunta a clientes.cliente_id
#   cantidad y monto → NOT NULL
#   Se almacenan medicamento y categoria directamente para consultas simples sin JOIN
cursor.execute("""
CREATE TABLE ventas (
    venta_id    INTEGER  PRIMARY KEY,
    cliente_id  INTEGER  NOT NULL,
    medicamento TEXT     NOT NULL,
    categoria   TEXT     NOT NULL,
    ciudad      TEXT     NOT NULL,
    cantidad    INTEGER  NOT NULL,
    monto       REAL     NOT NULL,
    fecha       TEXT     NOT NULL,
    FOREIGN KEY (cliente_id) REFERENCES clientes(cliente_id)
);
""")

# ═══════════════════════════════════════════════════════════════════════════
# INSERTAR DATOS
# ═══════════════════════════════════════════════════════════════════════════

# 6 clientes
cursor.executemany("INSERT INTO clientes VALUES (?, ?, ?);", [
    (1, 'Ana Torres',    'Lima'),
    (2, 'Luis Quispe',   'Arequipa'),
    (3, 'María Ramos',   'Lima'),
    (4, 'Carlos Vega',   'Cusco'),
    (5, 'Sofía López',   'Trujillo'),
    (6, 'Diego Mendoza', 'Lima'),
])

# 6 medicamentos
cursor.executemany("INSERT INTO medicamentos VALUES (?, ?, ?, ?);", [
    (1, 'Amoxicilina',    'Antibiótico',       25.50),
    (2, 'Paracetamol',    'Analgésico',          8.00),
    (3, 'Ibuprofeno',     'Antiinflamatorio',   15.00),
    (4, 'Vitamina C',     'Vitamina',           12.00),
    (5, 'Omeprazol',      'Antiácido',          18.50),
    (6, 'Loratadina',     'Antialérgico',       22.00),
])

# 18 ventas
cursor.executemany("INSERT INTO ventas VALUES (?, ?, ?, ?, ?, ?, ?, ?);", [
    ( 1, 1, 'Paracetamol',   'Analgésico',       'Lima',      5,  40.00, '2026-03-01'),
    ( 2, 2, 'Amoxicilina',   'Antibiótico',      'Arequipa',  2,  51.00, '2026-03-01'),
    ( 3, 3, 'Ibuprofeno',    'Antiinflamatorio', 'Lima',      3,  45.00, '2026-03-02'),
    ( 4, 4, 'Vitamina C',    'Vitamina',         'Cusco',     4,  48.00, '2026-03-02'),
    ( 5, 5, 'Loratadina',    'Antialérgico',     'Trujillo',  2,  44.00, '2026-03-03'),
    ( 6, 6, 'Omeprazol',     'Antiácido',        'Lima',      3,  55.50, '2026-03-03'),
    ( 7, 1, 'Amoxicilina',   'Antibiótico',      'Lima',      1,  25.50, '2026-03-04'),
    ( 8, 2, 'Vitamina C',    'Vitamina',         'Arequipa',  6,  72.00, '2026-03-04'),
    ( 9, 3, 'Paracetamol',   'Analgésico',       'Lima',     10,  80.00, '2026-03-05'),
    (10, 4, 'Loratadina',    'Antialérgico',     'Cusco',     3,  66.00, '2026-03-05'),
    (11, 5, 'Ibuprofeno',    'Antiinflamatorio', 'Trujillo',  2,  30.00, '2026-03-06'),
    (12, 6, 'Omeprazol',     'Antiácido',        'Lima',      4,  74.00, '2026-03-06'),
    (13, 1, 'Loratadina',    'Antialérgico',     'Lima',      5, 110.00, '2026-03-07'),
    (14, 2, 'Paracetamol',   'Analgésico',       'Arequipa',  8,  64.00, '2026-03-07'),
    (15, 3, 'Amoxicilina',   'Antibiótico',      'Lima',      3,  76.50, '2026-03-08'),
    (16, 4, 'Vitamina C',    'Vitamina',         'Cusco',     5,  60.00, '2026-03-08'),
    (17, 5, 'Ibuprofeno',    'Antiinflamatorio', 'Trujillo',  4,  60.00, '2026-03-09'),
    (18, 6, 'Paracetamol',   'Analgésico',       'Lima',     12,  96.00, '2026-03-09'),
])

conexion.commit()

# ── Función de visualización ──────────────────────────────────────────────────
# Solo se usa para MOSTRAR resultados de consultas SQL.
# NO reemplaza SQL para filtrar ni calcular.
def mostrar(consulta_sql):
    return pd.read_sql_query(consulta_sql, conexion)

print("Base de datos 'farmacia.db' creada correctamente.")
print("Tablas: clientes (6 filas) | medicamentos (6 filas) | ventas (18 filas)")
````
