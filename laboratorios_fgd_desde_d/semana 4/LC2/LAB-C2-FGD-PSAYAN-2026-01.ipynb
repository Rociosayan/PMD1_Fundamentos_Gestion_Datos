### Paso 3 — INNER JOIN: Reporte completo de ventas

El gerente necesita un reporte que muestre, en una sola tabla, el nombre del cliente, el producto comprado, el local y el monto de cada venta.

```python
# Reporte completo de ventas con INNER JOIN
df_reporte = pd.read_sql_query(
    '''
    SELECT
        v.id_venta,
        v.fecha,
        c.nombre_cliente,
        p.nombre_producto,
        l.nombre_local,
        v.cantidad,
        p.precio_unitario,
        ROUND(v.cantidad * p.precio_unitario, 2) AS total_venta
    FROM ventas v
    INNER JOIN clientes  c ON v.id_cliente  = c.id_cliente
    INNER JOIN productos p ON v.id_producto = p.id_producto
    INNER JOIN locales   l ON v.id_local    = l.id_local
    ORDER BY v.fecha DESC
    LIMIT 15
    ''',
    conn
)
display(df_reporte)
```

### Paso 4 — LEFT JOIN: Clientes sin compras

El área de marketing quiere identificar a los clientes registrados que nunca han realizado una compra, para enviarles una campaña de activación.

```python
# Todos los clientes y cuantas compras ha realizado cada uno
df_left = pd.read_sql_query(
    '''
    SELECT
        c.nombre_cliente,
        COUNT(v.id_venta) AS total_compras
    FROM clientes c
    LEFT JOIN ventas v ON c.id_cliente = v.id_cliente
    GROUP BY c.id_cliente
    ORDER BY total_compras ASC
    ''',
    conn
)
display(df_left)
```

### Paso 5 — GROUP BY: Ventas por categoría

La administradora quiere saber cuánto vende el minimarket por categoría de producto para planificar el abastecimiento del mes.

```python
# Total de ventas e ingresos por categoria
df_cat = pd.read_sql_query(
    '''
    SELECT
        cat.nombre_categoria,
        COUNT(v.id_venta) AS cantidad_ventas,
        ROUND(SUM(v.cantidad * p.precio_unitario), 2) AS ingresos_totales
    FROM ventas v
    INNER JOIN productos p ON v.id_producto = p.id_producto
    INNER JOIN categorias cat ON p.id_categoria = cat.id_categoria
    GROUP BY cat.id_categoria, cat.nombre_categoria
    ORDER BY ingresos_totales DESC
    ''',
    conn
)
display(df_cat)
```

### Paso 6 — HAVING: Solo locales con ventas significativas

El gerente general quiere analizar únicamente los locales que han registrado más de 20 ventas, para enfocar los esfuerzos de expansión en los más activos.

```python
# Locales con mas de 3 ventas registradas
df_having = pd.read_sql_query(
    '''
    SELECT
        l.nombre_local,
        l.distrito_local,
        COUNT(v.id_venta) AS cantidad_ventas,
        ROUND(SUM(v.cantidad * p.precio_unitario), 2) AS ingresos_totales
    FROM ventas v
    INNER JOIN locales l ON v.id_local = l.id_local
    INNER JOIN productos p ON v.id_producto = p.id_producto
    GROUP BY l.id_local, l.nombre_local, l.distrito_local
    HAVING cantidad_ventas > 3
    ORDER BY ingresos_totales DESC
    ''',
    conn
)
display(df_having)
```

El dueño del minimarket quiere saber cuáles son los 5 productos que generan más ingresos, considerando la cantidad vendida y el precio. Con esa información, decidirá qué productos potenciar.

```python
# Top 5 productos por ingresos totales
df_top = pd.read_sql_query(
    '''
    SELECT
        p.nombre_producto,
        cat.nombre_categoria,
        COUNT(v.id_venta) AS veces_vendido,
        SUM(v.cantidad) AS unidades_totales,
        ROUND(SUM(v.cantidad * p.precio_unitario), 2) AS ingresos_totales
    FROM ventas v
    INNER JOIN productos p ON v.id_producto = p.id_producto
    INNER JOIN categorias cat ON p.id_categoria = cat.id_categoria
    GROUP BY p.id_producto, p.nombre_producto, cat.nombre_categoria
    ORDER BY ingresos_totales DESC
    LIMIT 5
    ''',
    conn
)
display(df_top)
```
