## Actividad 4. Tratamiento de outliers


```python
# Paso 4.1 — Seleccionar variables numéricas relevantes
# cols_num = ['col1', 'col2', 'col3']  # ajustar al caso PMD1
cols_num = df_clean.select_dtypes(include='number').columns.tolist()
print('Variables numéricas detectadas:', cols_num)
```

```python
# Paso 4.2 — Boxplots antes del tratamiento
fig, axes = plt.subplots(1, len(cols_num), figsize=(5*len(cols_num), 4))
if len(cols_num) == 1: axes = [axes]
for ax, col in zip(axes, cols_num):
    df_clean[col].plot.box(ax=ax, title=col)
plt.suptitle('Boxplots antes del tratamiento de outliers')
plt.tight_layout()
plt.show()
```

```python
# Paso 4.3 — Calcular límites IQR
for col in cols_num:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lim_inf = Q1 - 1.5 * IQR
    lim_sup = Q3 + 1.5 * IQR
    n_out = ((df_clean[col] < lim_inf) | (df_clean[col] > lim_sup)).sum()
    print(f'{col}: Q1={Q1:.2f}, Q3={Q3:.2f}, IQR={IQR:.2f}, Límites=[{lim_inf:.2f}, {lim_sup:.2f}], Outliers={n_out}')
```

```python
# Paso 4.4 — Decisión y tratamiento (clip / eliminar / transformar)
# Ajustar según decisión de negocio del grupo
df_out = df_clean.copy()
for col in cols_num:
    Q1 = df_out[col].quantile(0.25)
    Q3 = df_out[col].quantile(0.75)
    IQR = Q3 - Q1
    lim_inf = Q1 - 1.5 * IQR
    lim_sup = Q3 + 1.5 * IQR
    # Opción clip (capado): mantiene el registro pero limita el valor extremo
    df_out[col] = df_out[col].clip(lower=lim_inf, upper=lim_sup)
print('Outliers tratados. Shape resultante:', df_out.shape)
```

```python
# Paso 4.5 — Comparación estadísticas descriptivas antes / después
print('=== ANTES ===')
print(df_clean[cols_num].describe().round(2))
print('\n=== DESPUÉS ===')
print(df_out[cols_num].describe().round(2))
```

## Actividad 5. Normalización de variables numéricas


```python
# Paso 5.1 — Seleccionar variables a normalizar (mínimo 3)
# vars_norm = ['col1', 'col2', 'col3']  # ajustar al caso PMD1
vars_norm = cols_num[:3] if len(cols_num) >= 3 else cols_num
print('Variables a normalizar:', vars_norm)
```

```python
# Paso 5.2 — Aplicar Min-Max normalización
# Cambiar a Z-score si la distribución es aproximadamente normal:
# df_norm[col + '_zscore'] = (df_out[col] - df_out[col].mean()) / df_out[col].std()

df_norm = df_out.copy()
for col in vars_norm:
    min_v = df_norm[col].min()
    max_v = df_norm[col].max()
    df_norm[col + '_norm'] = (df_norm[col] - min_v) / (max_v - min_v)

print('Normalización Min-Max aplicada.')
```

```python
# Paso 5.3 — Verificar rango [0, 1]
cols_n = [c + '_norm' for c in vars_norm]
print(df_norm[cols_n].describe().round(4))
```
